# Top scoring players through half of week 17

```python
import glob

import altair as alt
import pandas as pd
```

```python
lineups = pd.concat([pd.read_csv(f) for f in glob.glob('data/lineups_week_*.csv')]).sort_values('points', ascending=False)

cols = ['week', 'player_name', 'pos', 'team', 'fantasy_team_name', 'opponent_team_name', 'projected_points', 'points']
top_ten = lineups[cols].head(10)

top_ten
```

```python
alt.Chart(top_ten).mark_bar().encode(x='sum(points)', y='fantasy_team_name').properties(title='Most points for an owner out of the top ten players per week')
```

```python
alt.Chart(top_ten).mark_bar().encode(x='sum(points)', y='opponent_team_name').properties(title='Most points AGAINST an owner out of the top ten players per week')
```
